## Importing Preciptation data. Data includes county level precipitation data by year. It was collected by NOAA's National Centers for Environmental Information. Here, I analyze only the years between 2004 to 2019.

```python
import pandas as pd

df = pd.read_csv('Data/PrecipitationData - All.csv')
df.head()
```

## I create a new dataframe for each county-- Imperial, San Diego, and Riverside-- by querying my originial dataset. Next, I inspect the first few lines of the new dataframe to ensure this process worked correctly.

```python
Imp = df[df['County'] == 'Imperial'] 
Imp.head()
```

```python
SD = df[df['County'] == 'San Diego'] 
SD.head()
```

```python
Riv = df[df['County'] == 'Riverside '] 
Riv.head()
```

## Now that I have 3 seperate dataframes, I can plot each of them as a line on the graph below.

### The graph reveals the overall trends in precipitation over time. Although there is a slight increase in 2019, in general terms, the overall preciptation in inches is decreasing overtime for all three counties. 

```python
import matplotlib.pyplot as plt

plt.plot (Imp['Date'],Imp['Total P'], label= 'Imperial County')
plt.plot (Riv['Date'],Riv['Total P'], label= 'Riverside County')
plt.plot (SD['Date'],SD['Total P'], label= 'San Diego County')
plt.xlabel('Year') 
plt.ylabel('Total Precipitation (in)') 
plt.title("Total Precipitation over Time")
plt.legend()
plt.savefig('Precipitation.png')
```
