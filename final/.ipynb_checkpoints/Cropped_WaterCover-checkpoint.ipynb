```python
import rioxarray as rxr
import earthpy as et
import earthpy.spatial as es
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib.patches import Patch
import contextily as ctx
```

```python
years = [2004, 2019] # create a list of the comparison years to loop over
data = {} # Create an empty dictionary to save the raster dataset to 
data_masked = {} # Create empty dictionary to save masked numpy array to
dst_crs = 'EPSG:4326' 
```

```python
#Loop through years specified to create data needed to plot
for year in years:
    in_path = f'data/NLCD/NLCD_{str(year)}_Land_Cover_L48_20210604_gyGbthZM9gweVLuKaW5e.tiff'
    data[year] = rxr.open_rasterio(in_path, masked=True)# open raster and create element in 'data' dict.
    data[year] = data[year].rio.reproject(dst_crs)
    data_masked[year] = data[year].where(data[year]==11) # mask raster to show only were data is '11' (Open Water)
```

```python
data_masked[2004].plot(cmap=ListedColormap(['blue']))
```

```python
plt.imshow(data_masked[years[1]], extent=, cmap=ListedColormap(['blue']), zorder= 15) 
```

```python
fig, ax = plt.subplots(figsize=(15,10)) # create a figure and axis object to map onto and set the figure size.
data_masked[years[0]].plot(ax=ax, cmap=ListedColormap(['blue']), zorder= 15) #plot data, set extent to geography, set color and drawing order
data_masked[years[0]].plot(ax=ax, cmap=ListedColormap(['red']), zorder= 10)

# creates legend items
handles = [
    Patch(color='red', label='2004 Water Area'),
    Patch(color='blue', label='2019 Water Area')
]

ctx.add_basemap(ax, crs=dst_crs, source=ctx.providers.Esri.WorldImagery, zorder=5)
#ax.set_xlim([-116.29, -115.36])
#ax.set_ylim(ylim)
#plt.xlim([-116.29, -115.36])
#plt.axis('off')
plt.title('2004 and 2019 Open Water Area')
ax.legend(handles=handles)

plt.show()
```
