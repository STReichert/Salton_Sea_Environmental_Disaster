### Land Cover Mapping

One of our key questions for our research is the change in land uses over time, namely how agriculture has changed in the area. To explore this I look to the National Land Cover Database (NLCD) provided by the USGS and created from Landsat imagery. This data is provided as a raster file for various years in the past two decades. In this notebook I aim to explore the `rasterio` package to map land cover over a few years.

The raster files used to make these maps are too large to push to our github repository, but similar files can be downloaded from https://www.mrlc.gov/viewer/

```python
#import necessary packages
# Rasterio documentation can be found at https://rasterio.readthedocs.io/en/latest/

import numpy as np
import rasterio
import pandas as pd

# import plotting libraries
import matplotlib.pyplot as plt
import earthpy as et
import earthpy.plot as ep
import contextily as ctx
```

```python
# Pull raster file into a new object
dataset_2019 = rasterio.open('data/NLCD/NLCD_2019_Land_Cover_L48_20210604_gyGbthZM9gweVLuKaW5e.tiff')
```

Now that we've pulled the 2019 land cover data into our notebook we can begin exploring it, but what class is this object and what exactly can we do with it. We can use the python help function to show some more details about what can we do with this object.

```python
help(dataset_2019)
```

This is a DatasetReader class and we can see more about what we can do with it. We can also look to the documentation to explore the data more thoroughly.

Upon looking through the documentation you can find that this object reads the raster file, which contains data in the form of georeferenced pixels that contain data. Sometimes this data may come in a number of bands that contain different information, but some datasets only have 1 band. How many bands does NLCD data have?

```python
# count the number of bands in the raster object
print('The nubmer of bands in this dataset is:', dataset_2019.count)
```

This one band includes a number for each category of land cover analyzed in the database. The data dictionary can be found in another file in folder downloaded from USGS

```python
data_dict = pd.read_csv('data/NLCD/NLCD_landcover_legend_2018_12_17_gyGbthZM9gweVLuKaW5e.csv')
data_dict.dropna(axis=0, inplace=True) # Drop values with no legend information
data_dict # show legend
```

```python
legend = dict(zip(data_dict['Value'], data_dict['Legend'])) # save legend as a dictionary for plotting later.
```

Now that we have a data dictionary, what does the actual data look like? Underneatht the DatasetReader class is a numpy array for a given band in the file that we can read as shown below.

```python
dataset_2019.read(1) #reads the raster file in the first (in this case, only) band
```

Now that we can see the data in its abstract form, we can try to plot the data to show the categories.

```python
fig, ax = plt.subplots(figsize=(15,10)) # create a figure and axis object to map onto and set the figure size.
plt.colorbar(ax.imshow(dataset_2019.read(1), cmap='Paired')) #use the imshow method to plot the raster and wrap with the colorbar function to display a colorbar
plt.show()
```

We can see the Salton Sea in the center of the map, but it is hard to tell what the data is actually telling us. We could use the color bar on the side and cross reference it with the dataframe or dictionary we created earlier, but there must be a way for python to do it for us?

There is a why that relies on the earthpy library (insights from https://www.earthdatascience.org/courses/use-data-open-source-python/intro-raster-data-python/raster-data-processing/classify-plot-raster-data-in-python/)

To do this we 
- Find the unique data in the raster (because not all categories are necessarily present in this file)
- Map the categories in our legend to the values in this list of unique data to create a list of categories present in the data
- Plot the figure as before with an additional line to draw the legend.

```python
unique_data = np.unique(dataset_2019.read(1)) # Create list of unique values in our data
labels = list(np.vectorize(legend.get)(unique_data)) # Map legend categories onto this list of unique data, returning list of categories present in our data
```

```python
fig, ax = plt.subplots(figsize=(15,10)) # set up figure and axis objects and specify figure size
im = ax.imshow(dataset_2019.read(1), cmap='Paired') # plot the data on the axis object
ep.draw_legend(im, titles=labels) # use earthpy to draw a legend on the figure
plt.title('Land Cover Near the Salton Sea: 2019')
plt.show
```

This is vastly more readable! From here we can use the same code to explore land cover in different years as well as delve into creating custom color maps to even more clearly show the data
